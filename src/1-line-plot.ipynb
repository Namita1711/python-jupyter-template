### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt  # Import Matplotlib

# Step 1: Define x values (0 to 10)
x = list(range(11))  # [0, 1, 2, ..., 10]

# Step 2: Compute y values (square of x)
y = [i**2 for i in x]  # [0, 1, 4, 9, ..., 100]

# Step 3: Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label="y = x²")

# Step 4: Customize the plot
plt.title("Line Plot of y = x²")  # Add a title
plt.xlabel("X Values")  # Label for X-axis
plt.ylabel("Y Values (Squared)")  # Label for Y-axis
plt.grid(True)  # Add grid
plt.legend()  # Show legend

# Step 5: Display the plot
plt.show()
```
